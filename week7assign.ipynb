```python
import matplotlib.pyplot as plt

# Sample line graph
x = ["A", "B", "C", "D", "E", "F"]
y = [10, 5, 8, 12, 7, 10]

# Create a line plot
plt.plot(x, y)
plt.title("Commodities sold")
plt.xlabel("Commodity")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
# Sample bar graph
Product = ['Widget A', 'Widget B', 'Widget C']
Revenue = [250, 150, 200]

plt.bar(Product, Revenue, color='purple')
plt.title("Revenue Bar Graph")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
# Example: showing values in a histogram
Revenue = [250, 150, 200, 360, 210, 300]

plt.hist(Revenue, bins=5, color='grey')
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```
